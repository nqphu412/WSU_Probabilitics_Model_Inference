## Task 2:
You go for your yearly checkup and have several lab tests performed. A week later your doctor calls you and says she has good and bad news. The bad news is that you tested positive for a marker of a serious disease, and that the test is 99% accurate (i.e., the probability of testing positive given that you have the disease is 0.99, as is the probability of testing negative given that you do not have the disease). The good news is that this is a rare disease, striking only 1 in 20,000 people. Why is it good news that the disease is rare? What are the chances that you actually have the disease?

The probability of testing positive given that you have the disease: $P(Positive | Disease) = 0.99$
- which means $P(Positive | No Disease) = 1 - 0.99 = 0.01$

A rare disease, striking only 1 in 20,000 people: $P(Disease) = 1/20000$
- which means $P(No Disease) = 1 - 1/20000 = 19999/20000$

We have 
- $P(Positive) = P(Positive | Disease) \times P(Disease) + P(Positive | No Disease) \times P(No Disease)$
- $P(Positive) = 0.99 \times 1/20000 + 0.01 \times 19999/20000 = 0.010049$

Reason why it's a good news,
- $P(Disease | Positive) = P(Positive | Disease) \times P(Disease) / P(Positive)$
- $P(Disease | Positive) = 0.99 \times 1/20000 / 0.010049 = 0.0049 = 0.5\%$

It means even if the patients have a positive in the test, they only have 0.5% of actually having disease.
